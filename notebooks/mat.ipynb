```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y, 'r')
```

```python
# making a line plot
plt.plot(x, y, 'r')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')

plt.show()
```

```python
plt.subplot(2,1,1)
plt.ylabel('Y Label 1')
plt.xlabel('X Label 1')
plt.title('Title 1')
plt.plot(x, y, 'b')

plt.show()
```

```python
plt.subplot(1,2,2)
plt.plot(x, y, 'g')
```

```python
# object oriented approach

fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

ax.plot(x, y, 'r')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_title('Title')
```
